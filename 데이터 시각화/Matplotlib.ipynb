```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib

df=pd.read_csv(r'.\data\통합데이터.csv',encoding='utf-8',index_col='구별')
display(df.head())
```

```python
df_sorted=df.sort_values(by='CCTV 수량',ascending=False)#내림차순 정렬
display(df_sorted.head())
```

```python
import os
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import pandas as pd

# ① 절대 경로 권장 (상대경로 이슈 제거)
font_path = os.path.abspath(r"C:\Users\paint\OneDrive\Desktop\Githubproject\DataAnalysis\데이터 시각화\data\SB 어그로 L.ttf")

# ② 폰트를 매트플롯립 폰트 매니저에 '등록'
fm.fontManager.addfont(font_path)

# ③ 내부 이름 얻기 (여기에 나온 이름이 실제 family 이름)
family = fm.FontProperties(fname=font_path).get_name()
print("등록된 내부 이름:", family)  # 예: SB Aggro

# ④ 전역 설정 (리스트로 넣는 걸 추천)
mpl.rcParams.update({
    'font.family': [family],
    'axes.unicode_minus': False
})
```

```python
plt.style.use('bmh')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 깨짐 방지
fig,ax=plt.subplots(figsize=(12,5))
ax.bar(df_sorted.index,df_sorted['CCTV 수량'])

ax.set_title('구별 CCTV 수량')
ax.set_xlabel('구별')  
ax.set_ylabel('CCTV 수량')
ax.tick_params(axis='x',rotation=90)#x축 라벨 90도 회전
ax.grid(axis='y',linestyle=':',alpha=0.7)
plt.show()
```
