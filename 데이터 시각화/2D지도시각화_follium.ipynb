```python
import pandas as pd
geo_data=pd.read_json(r'.\data\N3A_G0100000.json')
mapo_features = [f for f in geo_data['features'] if f['properties'].get('NAME') == '마포구']
```

```python
mapo_features
```

```python
feature_collection = {
    "type": "FeatureCollection",
    "features": mapo_features  # ← 질문에 올려주신 리스트(마포구 1개)
}
```

```python
m = folium.Map(location=[37.5665, 126.9780], zoom_start=12, tiles='cartodbpositron')

layer = folium.GeoJson(
    data=feature_collection,
    name="마포구",
    style_function=lambda x: {"fillColor": "#3186cc", "color": "#225ea8", "weight": 2, "fillOpacity": 0.3},
    highlight_function=lambda x: {"weight": 3, "color": "#ff7800", "fillOpacity": 0.5},
    tooltip=folium.GeoJsonTooltip(fields=["NAME"], aliases=["구 이름"])  # ← 여기서 NAME 사용!
).add_to(m)

m
```

```python
import pandas as pd
df = pd.read_csv(r'.\data\통합데이터.csv')
```

```python
df['위도']=df['구별'].map(lambda x:seoul_gu.get(x,[0,0])[0])
df['경도']=df['구별'].map(lambda x:seoul_gu.get(x,[0,0])[1])
df.head()
```

```python
seoul_people_map=folium.Map(
    location=city_hall,
    zoom_start=11,
    tiles='cartodbpositron'
)

for idx,row in df.iterrows():
    folium.CircleMarker(
        location=[row['위도'],row['경도']],
        radius=row['인구수']/50000,
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
        popup=folium.Popup(f"{row['구별']}<br>인구수: {row['인구수']}", max_width=200)
    ).add_to(seoul_people_map)

seoul_people_map
```
