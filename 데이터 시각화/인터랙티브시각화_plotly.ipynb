```python
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
```

```python
pio.renderers.default ='notebook_connected'

df= px.data.iris()

fig=px.scatter(
    df,
    x='sepal_width',
    y='sepal_length',
    color='species',
    size='petal_length',
    hover_data=['sepal_width','sepal_length'], #마우스 오버 시 표시할 데이터
    width= 650,
    height=500,
    title='Scatter Plot of Iris Data'
)
fig.show()
```

```python
import plotly.express as px

df=px.data.gapminder() #데이터 필터 시에 .query로 필터 걸 수 잇음
fig=px.scatter(
    df,
    x='gdpPercap',
    y='lifeExp',
    size='pop',
    size_max=60,
    color='continent',
    log_x=True,
    range_y=[25,90],
    animation_frame='year',
    animation_group='country',
    title='Gapminder Animation of Life Expectancy vs GDP per Capita',
    width=650,
    height=500
)

fig.show()
```

```python
import plotly.express as px

df_2007=px.data.gapminder(year=2007) #2007년 데이터만 추출
fig=px.scatter(
    df_2007,
    x='gdpPercap',
    y='lifeExp',
    size='pop',
    size_max=60,
    color='continent',
    log_x=True,
    range_y=[25,90],
    title='Gapminder2007: Life Expectancy vs GDP per Capita',
    width=650,
    height=500
)

fig.update_layout(
    xaxis=dict(rangeslider_visible=True)
)

fig.show()
```

```python
import pandas as pd
df_apt=pd.read_csv('.\data\아파트 매매 실거래가_2020_2024.csv')
```

```python
df_apt.info()
```

```python
df_apt = df_apt.rename(columns={
    'aptDong': '법정동',
    'aptNm': '단지명',
    'buildYear': '건축년도',
    'buyerGbn': '구매자구분',
    'cdealDay': '계약일자',
    'cdealType': '계약유형',
    'dealAmount': '거래금액',
    'dealDay': '거래일',
    'dealMonth': '거래월',
    'dealYear': '거래연도',
    'dealingGbn': '거래구분',
    'estateAgentSggNm': '공인중개사소재지',
    'excluUseAr': '전용면적',
    'floor': '층',
    'jibun': '지번',
    'landLeaseholdGbn': '토지임대부구분',
    'rgstDate': '등록일자',
    'sggCd': '지역코드',
    'slerGbn': '매도자구분',
    'umdNm': '구별'
})

# 예시로 '가격' 컬럼을 새로 계산하고 싶다면 (거래금액이 문자열일 경우):
df_apt['가격'] = (
    df_apt['거래금액']
    .astype(str)
    .str.replace(',', '')  # 쉼표 제거
    .astype(float)
)
```

```python
# 1) 서울 자치구 코드 → 구 이름 매핑 (행정표준코드 5자리)
sgg_to_gu = {
    11110: "종로구",
    11140: "중구",
    11170: "용산구",
    11200: "성동구",
    11215: "광진구",
    11230: "동대문구",
    11260: "중랑구",
    11290: "성북구",
    11305: "강북구",
    11320: "도봉구",
    11350: "노원구",
    11380: "은평구",
    11410: "서대문구",
    11440: "마포구",
    11470: "양천구",
    11500: "강서구",
    11530: "구로구",
    11545: "금천구",
    11560: "영등포구",
    11590: "동작구",
    11620: "관악구",
    11650: "서초구",
    11680: "강남구",
    11710: "송파구",
    11740: "강동구",
}

# 2) 지역코드 dtype 정리(문자 → 정수)
#    (이미 int라면 astype(int)에서 변화 없습니다)
df_apt["지역코드"] = (
    df_apt["지역코드"]
    .astype(str)
    .str.extract(r"(\d+)")[0]      # 숫자만 추출
    .astype(float)                 # 결측/이상치 대비
    .astype("Int64")               # pandas nullable int
)

# 3) 매핑 적용: 구별 컬럼을 지역코드 기준으로 재설정
df_apt["구별"] = df_apt["지역코드"].map(lambda x: sgg_to_gu.get(int(x)) if pd.notna(x) else pd.NA)

# 4) 매핑되지 않은 코드 확인(있다면 리스트로 출력해 원인 파악)
missing_codes = (
    df_apt.loc[df_apt["구별"].isna() & df_apt["지역코드"].notna(), "지역코드"]
    .drop_duplicates()
    .tolist()
)
if missing_codes:
    print("매핑되지 않은 지역코드가 있습니다:", missing_codes)

# (선택) 만약 서울 외 지역 코드가 섞여 있거나, 기존 '구별'(읍/면/동명) 값을 유지하고 싶다면
# df_apt["구별"] = df_apt["구별"].fillna(df_apt["구별"])  # ← 이미 덮어썼으므로 불필요
# 또는 별도 컬럼으로 보관:
# df_apt["구별_from_code"] = df_apt["지역코드"].map(lambda x: sgg_to_gu.get(int(x)) if pd.notna(x) else pd.NA)
```
