```python
import pydeck as pdk

if '위도' not in df.columns or '경도' not in df.columns:
    raise ValueError("데이터프레임에 '위도' 및 '경도' 열이 포함되어 있어야 합니다.")
df=df.dropna(subset=['위도', '경도'])

df.info()
```

```python
#중복되는 값 제거
duplicates=df[(df['위도']==37.6398)&(df['경도']==127.0255)]
df=df.drop(duplicates.index[1:])  #첫 번째 항목을 제외한 중복 항목 제거
df.info()
```

```python
view_state=pdk.ViewState(
    latitude=df['위도'].mean(),
    longitude=df['경도'].mean(),
    zoom=10.5,
    min_zoom=8,
    max_zoom=13,
    #angle까지 정의 가능
    bearing=0,
    pitch=0 
)
```

```python
layer_scatter=pdk.Layer(
    'ScatterplotLayer',
    data=df,
    get_position='[경도, 위도]',
    get_color='[255,30,30,100]', #RGBA 색상(Red, Green, Blue, Alpha)
    get_radius=20, # 각 점의 반지름(미터단위)
)
```

```python
deck_scatter=pdk.Deck(
    layers=[layer_scatter],
    initial_view_state=view_state,
    map_style='light',
    tooltip={'text':'{안심주소}'}
)
```

```python
deck_scatter.to_html('3D_scatter.html',
                     notebook_display=False,
                     open_browser=False)
```

```python
# 1) 서울시 시군구 코드 → 구명 매핑 딕셔너리
SGG_CODE_TO_GU = {
    11110: "종로구",
    11140: "중구",
    11170: "용산구",
    11200: "성동구",
    11215: "광진구",
    11230: "동대문구",
    11260: "중랑구",
    11290: "성북구",
    11305: "강북구",
    11320: "도봉구",
    11350: "노원구",
    11380: "은평구",
    11410: "서대문구",
    11440: "마포구",
    11470: "양천구",
    11500: "강서구",
    11530: "구로구",
    11545: "금천구",
    11560: "영등포구",
    11590: "동작구",
    11620: "관악구",
    11650: "서초구",
    11680: "강남구",
    11710: "송파구",
    11740: "강동구",
}

df2["구별"] = df2["sggCd"].astype("Int64").map(SGG_CODE_TO_GU)
```

```python
import re
import numpy as np
import pandas as pd

# ----------------------------
# 0) 표준(타깃) 컬럼 정의
# ----------------------------
target_cols = [
    '법정동','지번','단지명','건축년도',
    '거래연도','거래월','거래일','지역코드',
    '층','전용면적','거래금액','구별','가격'
]

# ----------------------------
# 1) 원본(df2) → 표준 컬럼명 매핑
#    (아래 매핑에 없는 컬럼은 자동으로 버림)
# ----------------------------
rename_map = {
    'umdNm': '법정동',
    'jibun': '지번',
    'aptNm': '단지명',
    'buildYear': '건축년도',
    'dealYear': '거래연도',
    'dealMonth': '거래월',
    'dealDay': '거래일',
    'sggCd': '지역코드',
    'floor': '층',
    'excluUseAr': '전용면적',
    'dealAmount': '거래금액',
    # 시군구/자치구명 추정 컬럼 → '구별'
    # 필요시 여기에 더 추가
    # 'umdNm': '법정동',   # 상황에 따라 법정동으로 쓰고 싶다면 활성화
}

df_std = df2.rename(columns=rename_map).copy()

# ----------------------------
# 2) dealAmount(문자형) → 정수 변환(만원 가정)
#    - 쉼표/공백 제거
#    - 숫자 이외 문자 제거
# ----------------------------
if '거래금액' in df_std.columns:
    def to_int(x):
        if pd.isna(x):
            return np.nan
        s = str(x)
        s = s.replace(',', '').strip()
        # 숫자 이외 제거 (필요시 주석 처리)
        s = re.sub(r'[^0-9\-]', '', s)
        try:
            return int(s)
        except Exception:
            return np.nan

    df_std['거래금액'] = df_std['거래금액'].apply(to_int).astype('Int64')

# ----------------------------
# 3) 가격(억원) 파생: 거래금액(만원) 기준
#    - 부동산 실거래에서 흔히 거래금액이 "만원" 단위이므로 1억원=10,000만원
# ----------------------------
if '거래금액' in df_std.columns:
    df_std['가격'] = df_std['거래금액'] / 10000.0

# ----------------------------
# 4) 타입 정리
# ----------------------------
dtype_plan = {
    '법정동': 'string',
    '지번': 'string',
    '단지명': 'string',
    '건축년도': 'Int64',
    '거래연도': 'Int64',
    '거래월': 'Int64',
    '거래일': 'Int64',
    '지역코드': 'Int64',
    '층': 'Int64',
    '전용면적': 'float',
    '거래금액': 'Int64',
    '구별': 'string',
    '가격': 'float',
}

for c, t in dtype_plan.items():
    if c in df_std.columns:
        try:
            if t == 'string':
                df_std[c] = df_std[c].astype('string')
            elif t == 'Int64':
                # 안전 변환
                df_std[c] = pd.to_numeric(df_std[c], errors='coerce').astype('Int64')
            elif t == 'float':
                df_std[c] = pd.to_numeric(df_std[c], errors='coerce').astype(float)
        except Exception:
            pass

# ----------------------------
# 5) 표준 컬럼만 남기고, 존재 순서로 재정렬
# ----------------------------
existing = [c for c in target_cols if c in df_std.columns]
missing = [c for c in target_cols if c not in df_std.columns]

df_matched = df_std[existing].copy()

print(f"[정리 완료] 남은 컬럼: {len(existing)}개 / 표준 {len(target_cols)}개")
print("남긴 컬럼 순서:", existing)
if missing:
    print("누락된(원본에 없어서 삭제된) 표준 컬럼:", missing)
import numpy as np
import pandas as pd

# 예: df 안의 공백·빈칸을 NaN으로 바꾸기
df_matched = df_matched.replace(r'^\s*$', np.nan, regex=True)
df_matched = df_matched.dropna().reset_index(drop=True)

# 결과 확인
print(df_matched.info())
```

```python
import pandas as pd
import numpy as np
import pydeck as pdk

view_state_2d=pdk.ViewState(
    longitude=126.9979,
    latitude=37.5463,
    zoom=10.5,
    ptich=0
)

layer=pdk.Layer(
    'ScatterplotLayer',
    data=df_apt,
    get_position='[경도, 위도]',
    get_color='[200,30,0,160]',
    get_radius=100
)

deck=pdk.Deck(
    layers=[layer],
    initial_view_state=view_state_2d,
    map_style='light'
)

deck.to_html(
    'pydeck_apt_scatter.html',
    notebook_display=False,
    open_browser=False
)
```

```python
df_avg_price=df_apt.groupby('구별')['가격'].mean().reset_index()
df_avg_price['정규화가격']=df_avg_price['가격']/df_avg_price['가격'].max()
df_avg_price.head()
```

```python
import pandas as pd
gu_json=pd.read_json(r'.\data\N3A_G0100000.json')
df_gu_geo=pd.DataFrame()
#중구 데이터가 겹치므로 서울의 중구만 뽑아옴
seoul_features = [f for f in gu_json['features'] if f['properties']['BJCD'].startswith('11')]
# DataFrame 생성
df_gu_geo = pd.DataFrame({
    '구별': [f['properties']['NAME'] for f in seoul_features],
    '좌표': [f['geometry']['coordinates'] for f in seoul_features]
})
df_gu_geo.info()
```

```python
df_avg_price=df_avg_price.merge(
    df_gu_geo,
    on='구별',
    how='left'
)

df_avg_price['가격_문자']=df_avg_price['가격'].apply(lambda x:f'{x:.2f}')
df_avg_price.info()
```

```python
import pandas as pd
from copy import deepcopy

def explode_multipolygon(df, coord_col="좌표"):
    rows = []
    for _, r in df.iterrows():
        coords = r[coord_col]
        # MultiPolygon: coords[0][0][0] 가 list/tuple
        is_multi = isinstance(coords[0][0][0], (list, tuple))
        if is_multi:
            for poly in coords:  # 각 폴리곤 단위로 행 분리
                rr = deepcopy(r)
                rr[coord_col] = poly
                rows.append(rr)
        else:
            rows.append(r)
    return pd.DataFrame(rows, columns=df.columns)

df_plot = explode_multipolygon(df_avg_price)
```

```python
layer_polygon=pdk.Layer(
    'PolygonLayer',
    data=df_plot,
    get_polygon='좌표',
    opacity=0.8,
    get_fill_color='[0,255*정규화가격,0]',
    pickable=True,
    auto_highlight=True
)

tooltip={
    'html':'<b>{구별}</b><br/>평균가격: {가격_문자}억원',
    'style':{'color':'white','backgroundColor': 'steelblue'}}


deck_polygon=pdk.Deck(
    layers=[layer_polygon],
    initial_view_state=view_state_2d,
    map_style='dark',
    tooltip=tooltip
)

deck_polygon.to_html(
    'pydeck_apt_polygon_2d.html',
    notebook_display=False,
    open_browser=True
)
```
