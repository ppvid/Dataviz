```python
import pydeck as pdk

if '위도' not in df.columns or '경도' not in df.columns:
    raise ValueError("데이터프레임에 '위도' 및 '경도' 열이 포함되어 있어야 합니다.")
df=df.dropna(subset=['위도', '경도'])

df.info()
```

```python
#중복되는 값 제거
duplicates=df[(df['위도']==37.6398)&(df['경도']==127.0255)]
df=df.drop(duplicates.index[1:])  #첫 번째 항목을 제외한 중복 항목 제거
df.info()
```

```python
view_state=pdk.ViewState(
    latitude=df['위도'].mean(),
    longitude=df['경도'].mean(),
    zoom=10.5,
    min_zoom=8,
    max_zoom=13,
    #angle까지 정의 가능
    bearing=0,
    pitch=0 
)
```

```python
layer_scatter=pdk.Layer(
    'ScatterplotLayer',
    data=df,
    get_position='[경도, 위도]',
    get_color='[255,30,30,100]', #RGBA 색상(Red, Green, Blue, Alpha)
    get_radius=20, # 각 점의 반지름(미터단위)
)
```

```python
deck_scatter=pdk.Deck(
    layers=[layer_scatter],
    initial_view_state=view_state,
    map_style='light',
    tooltip={'text':'{안심주소}'}
)
```

```python
deck_scatter.to_html('3D_scatter.html',
                     notebook_display=False,
                     open_browser=True)
```

```python
df2=pd.read_csv('.\data\아파트 매매 실거래가_2020_2024.csv',encoding='utf-8')
```

```python
import re
import numpy as np
import pandas as pd

# ----------------------------
# 0) 표준(타깃) 컬럼 정의
# ----------------------------
target_cols = [
    '법정동','지번','단지명','건축년도',
    '거래연도','거래월','거래일','지역코드',
    '층','전용면적','거래금액','구별','가격'
]

# ----------------------------
# 1) 원본(df2) → 표준 컬럼명 매핑
#    (아래 매핑에 없는 컬럼은 자동으로 버림)
# ----------------------------
rename_map = {
    'umdNm': '법정동',
    'jibun': '지번',
    'aptNm': '단지명',
    'buildYear': '건축년도',
    'dealYear': '거래연도',
    'dealMonth': '거래월',
    'dealDay': '거래일',
    'sggCd': '지역코드',
    'floor': '층',
    'excluUseAr': '전용면적',
    'dealAmount': '거래금액',
    # 시군구/자치구명 추정 컬럼 → '구별'
    'estateAgentSggNm': '구별',
    # 필요시 여기에 더 추가
    # 'umdNm': '법정동',   # 상황에 따라 법정동으로 쓰고 싶다면 활성화
}

df_std = df2.rename(columns=rename_map).copy()

# ----------------------------
# 2) dealAmount(문자형) → 정수 변환(만원 가정)
#    - 쉼표/공백 제거
#    - 숫자 이외 문자 제거
# ----------------------------
if '거래금액' in df_std.columns:
    def to_int(x):
        if pd.isna(x):
            return np.nan
        s = str(x)
        s = s.replace(',', '').strip()
        # 숫자 이외 제거 (필요시 주석 처리)
        s = re.sub(r'[^0-9\-]', '', s)
        try:
            return int(s)
        except Exception:
            return np.nan

    df_std['거래금액'] = df_std['거래금액'].apply(to_int).astype('Int64')

# ----------------------------
# 3) 가격(억원) 파생: 거래금액(만원) 기준
#    - 부동산 실거래에서 흔히 거래금액이 "만원" 단위이므로 1억원=10,000만원
# ----------------------------
if '거래금액' in df_std.columns:
    df_std['가격'] = df_std['거래금액'] / 10000.0

# ----------------------------
# 4) 타입 정리
# ----------------------------
dtype_plan = {
    '법정동': 'string',
    '지번': 'string',
    '단지명': 'string',
    '건축년도': 'Int64',
    '거래연도': 'Int64',
    '거래월': 'Int64',
    '거래일': 'Int64',
    '지역코드': 'Int64',
    '층': 'Int64',
    '전용면적': 'float',
    '거래금액': 'Int64',
    '구별': 'string',
    '가격': 'float',
}

for c, t in dtype_plan.items():
    if c in df_std.columns:
        try:
            if t == 'string':
                df_std[c] = df_std[c].astype('string')
            elif t == 'Int64':
                # 안전 변환
                df_std[c] = pd.to_numeric(df_std[c], errors='coerce').astype('Int64')
            elif t == 'float':
                df_std[c] = pd.to_numeric(df_std[c], errors='coerce').astype(float)
        except Exception:
            pass

# ----------------------------
# 5) 표준 컬럼만 남기고, 존재 순서로 재정렬
# ----------------------------
existing = [c for c in target_cols if c in df_std.columns]
missing = [c for c in target_cols if c not in df_std.columns]

df_matched = df_std[existing].copy()

print(f"[정리 완료] 남은 컬럼: {len(existing)}개 / 표준 {len(target_cols)}개")
print("남긴 컬럼 순서:", existing)
if missing:
    print("누락된(원본에 없어서 삭제된) 표준 컬럼:", missing)

# 결과 확인
print(df_matched.info())
df_matched.head()
```

```python
unique_addresses=df_matched['주소지'].unique()

print(f"고유 주소지 개수: {len(unique_addresses)}개")
unique_addresses[:5]
```

```python
import os
if not os.path.exists('./data/아파트실거래가_위도경도.csv'):
    import requests
    apiurl = "https://api.vworld.kr/req/address?"

    geo_apt={}
    for address in unique_addresses:
        params = {
            "service": "address",
            "request": "getcoord",
            "crs": "epsg:4326",
            "address": address,
            "format": "json",
            "type": "road",
            "key": "3346A0C0-1596-3B19-903C-6CADB520EC4A"
        }
        response = requests.get(apiurl, params=params)
        if response.status_code == 200 and response.json()['response']['status']=='OK':
            point= response.json()['response']['result']['point']
            geo_apt[address]=(point['y'],point['x'])  #위도,경도
    print('처리 전 유일한 주소지의 개수:',len(unique_addresses), end='\r')
    print(f'처리된 주소지 개수: {len(geo_apt)}', end='\r')
    df_matched['위도']=df_matched['주소지'].map(lambda addr: geo_apt.get(addr,(np.nan,np.nan))[0])
    df_matched['경도']=df_matched['주소지'].map(lambda addr: geo_apt.get(addr,(np.nan,np.nan))[1])
    df_matched.to_csv('./data/아파트실거래가_위도경도.csv',index=False,encoding='utf-8')
    df_matched=pd.read_csv('./data/아파트실거래가_위도경도.csv',encoding='utf-8')
else:
    df_matched=pd.read_csv('./data/아파트실거래가_위도경도.csv',encoding='utf-8')
```
